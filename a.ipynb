1. Hiển thị toàn bộ dữ liệu

```python
import pandas as pd
data = pd.read_csv('BTH2.csv')
data
```

2. Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình

```python
data.isna().sum()
```

```python
age_mean = data['Age'].mean()
income_mean = data['Income'].mean()
values = {"Age": age_mean.astype(int), "Income": income_mean.astype(int)}
print(values)
data.fillna(value=values,inplace=True)
data
```

3. Hiển thị các bản ghi có Income > 70000 và Age > 50

```python
data2 = data[(data['Income'] > 70000) & (data['Age'] > 50)]
data2
```

4. Chuẩn hóa dữ liệu theo min max

```python
from sklearn import preprocessing as pp
mms = pp.MinMaxScaler()
data3 = data
data3[['Age', 'Income']] = mms.fit_transform(data3[['Age', 'Income']])
data3
```

5. Liên tục hóa thuộc tính Region

```python
from sklearn.preprocessing import LabelEncoder as le
data['Region Encoded'] = le().fit_transform(data['Region'])
data
```

8. Rời rạc hóa thuộc tính Income

```python
min = data['Income'].min()
max = data['Income'].max()
print(min)
print(max)
range = [60000,80000]
labels = ["<60000","60000-80000",">80000"]
data['Income'] = pd.cut(data['Income'], bins=[min,60000,80000,max],labels=labels, include_lowest=True)
data
```
